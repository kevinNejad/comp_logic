```python
import requests
```

```python
## Look https://github.com/commonsense/conceptnet5/wiki/Relations 
'''
IsA: car -> vehicle; Chicago -> city
PartOf: gearshit -> car
HasA: bird -> wing; pen -> ink
UsedFor: bridge -> cross water
CapableOf: knife -> cut
AtLocation: butter -> refrigerator; London -> UK
Causes: exercise -> sweat
HasProperty: ice -> cold
Desires: person -> love
CreatedBy: cake -> bake
CausesDesire: having no food -> go to store
MadeOf: bottle -> plastic
'''
relations = ['IsA','PartOf','HasA','UsedFor','CapableOf','AtLocation',
            'Causes','HasProperty','Desires','CreatedBy','CausesDesire','MadeOf']
```

```python
API_URI = 'http://api.conceptnet.io/'
EN = 'en'
def get_data(target, relations):
    data={}
    for rel in relations:
        query = 'query?start=/c/{}/{}&rel=/r/{}&limit=1000'.format(EN,target,rel)
        obj = requests.get(API_URI+query).json()
        
        # check if such a relation exist
        if len(obj['edges']) == 0:
            continue
        info = extract_info(obj)
        data[rel] = info
    return data

def extract_info(obj):
    labels = []
    for edge in obj['edges']:
        labels.append(edge['end']['label'])
        
    return labels
```

```python
data = get_data('teacher', relations)
```
